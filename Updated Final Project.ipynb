# CLEANING

```python
student_performance["Teacher_Quality"].fillna("Not Provided", inplace = True)
student_performance["Parental_Education_Level"].fillna("Not Provided", inplace = True)
student_performance["Distance_from_Home"].fillna("Not Provided", inplace = True)

student_performance.isna().sum()
```

```python
student_performance["Exam_Score"].replace({101: 100}, inplace = True)

student_performance.head(5)
```

# TRANSFORMATION

```python
student_performance.loc[student_performance["Hours_Studied"] >= 35, "Readers"] = "Good Readers"
student_performance.loc[(student_performance["Hours_Studied"] >= 20) & (student_performance["Hours_Studied"] < 35), "Readers" ] = "Average Readers"
student_performance.loc[(student_performance["Hours_Studied"] >= 10) & (student_performance["Hours_Studied"] < 20), "Readers" ] = "Fair Readers"
student_performance.loc[student_performance["Hours_Studied"] < 10, "Readers"] = "Poor Readers"

student_performance.head()
```

```python
student_performance.loc[student_performance["Attendance"] >= 85, "Attendance Status"] = "Outstanding"
student_performance.loc[(student_performance["Attendance"] >= 70) & (student_performance["Attendance"] < 85), "Attendance Status"] = "Good"
student_performance.loc[student_performance["Attendance"] < 70, "Attendance Status"] = "Needs Improvement"

student_performance.head()
```

```python
student_performance.loc[student_performance["Sleep_Hours"] >= 9, "Sleeping Status"] = "Above Normal"
student_performance.loc[(student_performance["Sleep_Hours"] >= 7) & (student_performance["Sleep_Hours"] < 9), "Sleeping Status"] = "Recommended"
student_performance.loc[student_performance["Sleep_Hours"] < 7, "Sleeping Status"] = "Insufficient"

student_performance.head()
```

```python
student_performance.loc[student_performance["Exam_Score"] >= 75, "Grading"] = "Excellent"
student_performance.loc[(student_performance["Exam_Score"] >= 65 ) & (student_performance["Exam_Score"] < 75), "Grading"] = "Good"
student_performance.loc[student_performance["Exam_Score"] < 65, "Grading"] = "Needs Improvement"

student_performance.head()
```

```python
student_performance.loc[student_performance["Previous_Scores"] >= 85, "Previous_Grading"] = "Excellent"
student_performance.loc[(student_performance["Previous_Scores"] >= 70 ) & (student_performance["Previous_Scores"] < 85), "Previous_Grading"] = "Good"
student_performance.loc[student_performance["Previous_Scores"] < 70, "Previous_Grading"] = "Needs Improvement"

student_performance.head()
```

```python
student_performance["Student_ID"] = range(1, 6608)
student_performance.head(5)
```

```python
grading_group = student_performance.groupby("Grading").agg({
    "Student_ID": "count",
    "Exam_Score": "mean",
    "Previous_Scores": "mean"})
print (grading_group)


grouped = student_performance.groupby("Grading")["Student_ID"].count() 

plt.pie(
    grouped,                                
    labels=grouped.index,                    
    autopct="%1.1f%%",
    colors=(["pink", "red", "orange"])                                               
) 
plt.title("Total numbers of Student By Grading")                        
plt.show()
```

```python
previous_grading_group = student_performance.groupby("Previous_Grading").agg({
    "Student_ID": "count",
    "Exam_Score": "mean",
    "Previous_Scores": "mean"})
print(previous_grading_group)

grouped = student_performance.groupby("Previous_Grading")["Student_ID"].count() 

plt.pie(
    grouped,                                
    labels=grouped.index,                    
    autopct="%1.1f%%",
    colors=(["pink", "red", "orange"])                                               
) 
plt.title("Total numbers of Student By Previous_Grading")                        
plt.show()
```

# GENERATING INSIGHTS

# 1. Understand Key Factors Influencing Exam Scores

```python
# How do study hours, attendance, and sleep affect exam performance?


R1 = student_performance['Exam_Score'].corr(student_performance['Attendance'])
R2 = student_performance['Exam_Score'].corr( student_performance["Hours_Studied"])
R3 = student_performance['Exam_Score'].corr( student_performance["Sleep_Hours"])

print("Cor:", R1)
print("Cor:", R2)
print("Cor:", R3)

Attendance_Status = student_performance.groupby("Attendance Status").agg({
    "Student_ID": "count",
    "Exam_Score": "mean"})
print(Attendance_Status)

Readers = student_performance.groupby("Readers").agg({
    "Student_ID": "count",
    "Exam_Score": "mean"})
print(Readers)


Sleeping_Status = student_performance.groupby("Sleeping Status").agg({
    "Student_ID": "count",
    "Exam_Score": "mean"})
print(Sleeping_Status)
```

```python
#Which factor contributes most to higher scores: motivation, previous scores, or tutoring?

Motivation_Level = student_performance.groupby("Motivation_Level").agg({
    "Student_ID": "count",
    "Exam_Score": "mean"})
print(Motivation_Level)



Tutoring_Sessions = student_performance.groupby("Motivation_Level").agg({
    "Student_ID": "count",
    "Exam_Score": "mean"})
print(Tutoring_Sessions)


previous_grading_group = student_performance.groupby("Previous_Grading").agg({
    "Student_ID": "count",
    "Exam_Score": "mean"})
print(previous_grading_group)
```

```python
#Do students with internet access perform better than those without?

Internet_Access = student_performance.groupby("Internet_Access").agg({
    "Student_ID": "count",
    "Exam_Score": "mean"})
print(Internet_Access)



plt.figure(figsize=(8, 4)) # you can alter the size of your chart
plt.bar(student_performance["Internet_Access"], student_performance["Exam_Score"].mean(), color="red")
plt.title("Internet_Access")
plt.xlabel("Internet_Access")
plt.ylabel("Total_No_Of_Students")
plt.show()
```
