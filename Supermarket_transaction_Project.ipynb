# Libraries

```python
import pandas as pd

import numpy as np

import seaborn as sns 

import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel(r"C:\Users\DELL\Downloads\supermarket_transactions.xlsx")
spt_copied = df.copy()
spt_copied
```

# Cleaning & Manipulation

```python
spt_copied.drop('Unnamed: 0', axis= 1, inplace=True)
```

```python
spt_copied['Year']= spt_copied['timestamp'].dt.year
spt_copied['Month']= spt_copied['timestamp'].dt.month_name()
spt_copied['Day']= spt_copied['timestamp'].dt.day_name()
spt_copied['quarter']= spt_copied['timestamp'].dt.quarter

spt_copied['Year']=spt_copied['Year'].astype('object')
spt_copied['quarter']=spt_copied['quarter'].astype('object')
```

# EDA

```python
print("\nDescriptive statistics of numerical variables:")



print(spt_copied[['quantity', 'total_amount', 'unit_price']].describe().round(2))
```

```python
#  Correlation Heatmap

print("\nCorrelation heatmap of numerical variables:")

numerical_cols = ['quantity', 'total_amount', 'unit_price']

plt.figure(figsize=(8, 4))
corr = spt_copied[numerical_cols].corr().round(2)
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(
    corr,
    mask=mask,
    annot=True,
    fmt='.2f',
    cmap='coolwarm',
    vmin=-1,
    vmax=1,
    center=0,
    linewidths=0.5
)

plt.title('Correlation Matrix - Supermarket Transaction', fontsize=10)
#plt.savefig('boston_correlation_heatmap.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
#   Boxplots - Outlier Detection 

print("\nBoxplot of Numerical variables:")

plt.figure(figsize=(20, 20))

plt.subplot(2, 2, 1)
sns.boxplot(y= spt_copied['quantity'])
plt.title('Boxplot of Supermarket Quantity Sold')

plt.subplot(2, 2, 2)
sns.boxplot(y= spt_copied['total_amount'])
plt.title('Boxplot of Supermarket total_amount')
```

# Visualization

```python
# Number of sales by product_name


print("\nNumber of sales by product_name:")

plt.figure(figsize=(20, 6))

product_name_cnt = spt_copied.groupby(
    'product_name', observed=False
    )['id'].count().sort_values(ascending=False).rename('Number of sales').round(2).head(10)

ax = sns.barplot(x=product_name_cnt.index, 
            y=product_name_cnt.values, 
            hue=product_name_cnt.index, 
            palette='Blues_r', 
            legend=False
)
plt.title('Number of sales by product_name')
plt.xlabel('product_name')
plt.ylabel('Number of sales')
plt.grid(axis='y', alpha=0.3)

for container in ax.containers:
    ax.bar_label(container, 
                 fmt='{:,.2f}',      
                 padding=3, 
                 fontsize=7,
                 fontweight='light'
                 )

plt.tight_layout()
plt.show()

product_name_cnt
```

```python
# quantity_sold VS total_revenue by product_name

fig, axes = plt.subplots(1, 2, figsize=(20, 6))


product_name_var = spt_copied.groupby(
    "product_name", observed=False).agg({
    "quantity": "sum",
    "total_amount": "sum"}).rename(columns={
        "quantity": "quantity_sold",
        "total_amount": "total_revenue"
    }).round(2)
    
qty_sorted = product_name_var.sort_values(by="quantity_sold", ascending=False).head(5)
rev_sorted = product_name_var.sort_values(by="total_revenue", ascending=False).head(5)
    
    

sns.barplot(
    data=qty_sorted.reset_index(),
    x="product_name", 
    y="quantity_sold", 
    hue="product_name", 
    palette='Oranges_d', 
    legend=False,
    ax=axes[0]
)
axes[0].set_title('quantity_sold by product_name (top 5)', fontsize=14, pad=15)
axes[0].set_xlabel('product_name')
axes[0].set_ylabel('quantity_sold')
axes[0].tick_params(axis='x')

# Adding value labels on top of bars
for container in axes[0].containers:
    axes[0].bar_label(container, fmt='%d', padding=3)
    


# Chart 2: total_revenue

sns.barplot(
    data=rev_sorted.reset_index(),
    x="product_name",
    y="total_revenue",
    hue="product_name",
    palette="Greens_d",
    legend=False,
    ax=axes[1]
)

axes[1].set_title('total_revenue by product_name (top 5)', fontsize=14, pad=15)
axes[1].set_xlabel('product_name')
axes[1].set_ylabel('total_revenue')
axes[1].tick_params(axis='x')

# Adding value labels on top of bars
for container in axes[1].containers:
    axes[1].bar_label(container, fmt='%.0f', padding=3)

plt.tight_layout()
plt.show()
    
```

```python
# Yearly sales trend


yearly_sales = spt_copied.groupby(
    'Year', observed=False
    )['quantity'].sum().rename('Total quantity sold').reset_index()


sns.set_style("whitegrid") # To create dashboard backgoround
plt.figure(figsize=(20, 6))


ax = sns.lineplot(data =yearly_sales, x = 'Year', y ='Total quantity sold',  marker = "o", linestyle = "--")

for x, y in zip(yearly_sales["Year"], yearly_sales["Total quantity sold"]):
    ax.text(x, y, f'{y:.2f}', ha='center', va='bottom')
    
plt.title("Yearly sales trend")
plt.xlabel("Year")
plt.ylabel("Total quantity sold")
plt.tight_layout()
plt.show()

yearly_sales
```

```python
# Yearly revenue trend


yearly_revenue = spt_copied.groupby(
    'Year', observed=False
    )['total_amount'].sum().rename('Revenue').reset_index()

sns.set_style("darkgrid") # To create dashboard backgoround
plt.figure(figsize=(20, 6))


ax = sns.lineplot(data =yearly_revenue, x = 'Year', y ='Revenue', color = "black", marker = "o", linestyle = "--")

for x, y in zip(yearly_revenue["Year"], yearly_revenue["Revenue"]):
    ax.text(x, y, f'{y:.2f}', ha='center', va='bottom')
    
plt.title("Yearly revenue trend")
plt.xlabel("Year")
plt.ylabel("Revenue")
plt.tight_layout()
plt.show()


yearly_revenue.round(2)
```

```python
# Monthly transaction


monthly_cust_cnt = spt_copied.groupby(
    'Month', observed=False
    )['id'].count().rename('Total_customers').reset_index()

month_order = ["January", "February", "March", "April", "May", "June", 
               "July", "August", "September", "October", "November", "December"]

monthly_cust_cnt["Month"] = pd.Categorical(
    monthly_cust_cnt["Month"],
    categories=month_order,
    ordered=True
)

monthly_cust_cnt = monthly_cust_cnt.sort_values("Month")

monthly_cust_cnt
```

```python
# Monthly sales trend


monthly_sales = spt_copied.groupby(
    ['Year', 'Month'], observed=False
    )['quantity'].sum().rename('Total_sales').reset_index()

month_order = ["January", "February", "March", "April", "May", "June", 
               "July", "August", "September", "October", "November", "December"]

monthly_sales["Month"] = pd.Categorical(
    monthly_sales["Month"],
    categories=month_order,
    ordered=True
)

monthly_sales = monthly_sales.sort_values("Month")



sns.set_style("darkgrid") # To create dashboard backgoround
plt.figure(figsize=(20, 6))


ax = sns.lineplot(
    data =monthly_sales, x = 'Month', y ='Total_sales', 
    hue = "Year", marker = "o", linestyle = "--", palette = 'tab10')

#for x, y in zip(monthly_sales["Month"], monthly_sales["Total_sales"]):
    #ax.text(x, y, f'{y:.2f}', ha='center', va='bottom')
    
plt.title("Monthly sales trend")
plt.xlabel("Month")
plt.ylabel("Total_sales")
plt.tight_layout()
plt.show()


monthly_sales.round(2)
```

```python
# Monthly revenue trend

monthly_revenue = spt_copied.groupby(
    ['Year', 'Month'], observed=False
    )['total_amount'].sum().rename('Revenue').reset_index()

month_order = ["January", "February", "March", "April", "May", "June", 
               "July", "August", "September", "October", "November", "December"]

monthly_revenue["Month"] = pd.Categorical(
    monthly_revenue["Month"],
    categories=month_order,
    ordered=True
)

monthly_revenue = monthly_revenue.sort_values("Month")



sns.set_style("darkgrid") # To create dashboard backgoround
plt.figure(figsize=(20, 7))


ax = sns.lineplot(
    data =monthly_revenue, x = 'Month', y ='Revenue', 
    hue = "Year", marker = "o", linestyle = "--", palette = 'tab10')

#for x, y in zip(monthly_revenue["Month"], monthly_revenue["Revenue"]):
    #ax.text(x, y, f'{y:.2f}', ha='center', va='bottom')
    
plt.title("Monthly revenue trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.legend(title="Year", title_fontsize=12, fontsize=10)
plt.tight_layout()
plt.show()


monthly_revenue.round(2)
```
