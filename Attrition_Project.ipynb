# Objective 1: Identify Workforce Demographics Most at Risk of Attrition

#1. Using pandas groupby, which demographic groups (age band, gender, department, role) have the highest attrition rate?

```python
Age_band = Attrition_Project_Copied[Attrition_Project_Copied["Attrition"]== "Yes"].groupby("Age Group")["EmployeeNumber"].count()
Age_band 
```

```python
Gender_role = Attrition_Project_Copied[Attrition_Project_Copied["Attrition"]== "Yes"].groupby("Gender")["EmployeeNumber"].count()
Gender_role
```

```python
Department = Attrition_Project_Copied[Attrition_Project_Copied["Attrition"]== "Yes"].groupby("Department")["EmployeeNumber"].count()
Department

```

```python
JobRole = Attrition_Project_Copied[Attrition_Project_Copied["Attrition"]== "Yes"].groupby("JobRole")["EmployeeNumber"].count()
JobRole
```

#2. What percentage of employees leave within their first year, by department and role?

```python
Dept = Attrition_Project_Copied[(Attrition_Project_Copied["Attrition"] =="Yes") & 
                                (Attrition_Project_Copied["YearsAtCompany"] == 1)].groupby("Department")["EmployeeNumber"].count()
Dept

plt.pie(
    Dept,                                
    labels=Dept.index,                   
    autopct="%1.1f%%",
    colors=(["blue", "red", "purple", "green"])                                               
) 
plt.title("Attrition By Dept In 1st Year")                        
plt.show()
```

```python
Jrole = Attrition_Project_Copied[(Attrition_Project_Copied["Attrition"] =="Yes") & 
                                   (Attrition_Project_Copied["YearsAtCompany"] == 1)].groupby("JobRole")["EmployeeNumber"].count()
Jrole

plt.pie(
    Jrole,                                
    labels=Jrole.index,                   
    autopct="%1.1f%%",
    colors=(["blue", "red", "purple", "green"])                                               
) 
plt.title("Attrition By JobRole In 1st Year")                        
plt.show()
```
